```python
def display_board(row1,row2,row3):
    from IPython.display import clear_output
    clear_output()
    print('-------------------------')
    print('|       |       |       |')
    print(''.join(row1))
    print('|       |       |       |')
    print('-------------------------')
    print('|       |       |       |')
    print(''.join(row2))
    print('|       |       |       |')
    print('-------------------------')
    print('|       |       |       |')
    print(''.join(row3))
    print('|       |       |       |')
    print('-------------------------')
```

```python
def position_choice(which_player,avail_num):
    choice = 'wrong'
    nums = ['1','2','3','4','5','6','7','8','9']
    while choice not in avail_num and which_player%2 != 0:
        choice = input("Player 1: Pick a square (1-9): ")
        if choice not in avail_num and choice in nums:
            print("Sorry, that square is already taken.")
        elif choice not in nums:
            print("Sorry, that choice is not within 1-9")
    while choice not in avail_num and which_player%2 == 0:
        choice = input("Player 2: Pick a square (1-9): ")
        if choice not in avail_num and choice in nums:
            print("Sorry, that square is already taken.")
        elif choice not in nums:
            print("Sorry, that choice is not within 1-9")
    avail_num.remove(choice)
    return int(choice)
```

```python
def replacement_choice(position,row1,row2,row3,which_player,players):
    if position == 1:
        if which_player%2 != 0:
            row3[4] = players['player_one']
        else:
            row3[4] = players['player_two']
    elif position == 2:
        if which_player%2 != 0:
            row3[12] = players['player_one']
        else:
            row3[12] = players['player_two']
    elif position == 3:
        if which_player%2 != 0:
            row3[20] = players['player_one']
        else:
            row3[20] = players['player_two']
    elif position == 4:
        if which_player%2 != 0:
            row2[4] = players['player_one']
        else:
            row2[4] = players['player_two']
    elif position == 5:
        if which_player%2 != 0:
            row2[12] = players['player_one']
        else:
            row2[12] = players['player_two']
    elif position == 6:
        if which_player%2 != 0:
            row2[20] = players['player_one']
        else:
            row2[20] = players['player_two']
    elif position == 7:
        if which_player%2 != 0:
            row1[4] = players['player_one']
        else:
            row1[4] = players['player_two']
    elif position == 8:
        if which_player%2 != 0:
            row1[12] = players['player_one']
        else:
            row1[12] = players['player_two']
    elif position == 9:
        if which_player%2 != 0:
            row1[20] = players['player_one']
        else:
            row1[20] = players['player_two']
```

```python
def win_check(row1,row2,row3):
    if row1[4] in ['X','O'] and row1[4] == row1[12] and row1[12] == row1[20]:
        return True
    elif row2[4] in ['X','O'] and row2[4] == row2[12] and row2[12] == row2[20]:
        return True
    elif row3[4] in ['X','O'] and row3[4] == row3[12] and row3[12] == row3[20]:
        return True
    elif row1[4] in ['X','O'] and row1[4] == row2[12] and row2[12] == row3[20]:
        return True
    elif row1[20] in ['X','O'] and row1[20] == row2[12] and row2[12] == row3[4]:
        return True
    elif row1[4] in ['X','O'] and row1[4] == row2[4] and row2[4] == row3[4]:
        return True
    elif row1[12] in ['X','O'] and row1[12] == row2[12] and row2[12] == row3[12]:
        return True
    elif row1[20] in ['X','O'] and row1[20] == row2[20] and row2[20] == row3[20]:
        return True
    else:
        return False
```

```python
def gameon_choice():
    choice = 'wrong'
    while choice not in ['Y','N']:
        choice = input("Would you like to play again? Type Y or N: ").upper()
        if choice not in ['Y','N']:
            print("Sorry, I don't understand. Please type Y or N: ")
    return choice == 'Y'
```

```python
def player_select(players):
    players['player_one'] = input("Welcome to Tic-Tac-Toe!\nPlayer 1: Would you like to be X or O? ").upper()
    while players['player_one'] not in ['X','O']:
        players['player_one'] = input("Invalid choice.\nPlayer 1: Please type X or O ").upper()
    if players['player_one'] == 'X':
        players['player_two'] = 'O'
    else:
        players['player_two'] = 'X'
```

```python
def tic_tac_toe():
    players = {'player_one':'v1','player_two':'v2'}
    which_player = rand_player()
    wincheck = False
    avail_num = ['1','2','3','4','5','6','7','8','9']
    row1 = (['|',' ',' ',' ',' ',' ',' ',' ','|',' ',' ',' ',' ',' ',' ',' ','|',' ',' ',' ',' ',' ',' ',' ','|'])
    row2 = (['|',' ',' ',' ',' ',' ',' ',' ','|',' ',' ',' ',' ',' ',' ',' ','|',' ',' ',' ',' ',' ',' ',' ','|'])
    row3 = (['|',' ',' ',' ',' ',' ',' ',' ','|',' ',' ',' ',' ',' ',' ',' ','|',' ',' ',' ',' ',' ',' ',' ','|'])
    player_select(players)
    print("Player {} will go first.".format(which_player))
    input("Press enter to begin.")
    while wincheck == False:
        if avail_num != []:
            display_board(row1,row2,row3)
            position = position_choice(which_player,avail_num)
            replacement_choice(position,row1,row2,row3,which_player,players)
            which_player += 1
            wincheck = win_check(row1,row2,row3)
        else:
            display_board(row1,row2,row3)
            print("Sorry, nobody wins.")
            break
    if wincheck:
        if which_player%2 != 0:
            display_board(row1,row2,row3)
            print("Player 2 wins!")
        else:
            display_board(row1,row2,row3)
            print("Player 1 wins!")
    game_on = gameon_choice()
    if game_on:
        tic_tac_toe()
    else:
        print('Goodbye!')
```

```python
def rand_player():
    import random
    return random.randint(1,2)
```
